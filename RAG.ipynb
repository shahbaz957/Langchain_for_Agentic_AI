```python
import os
from dotenv import load_dotenv
# Load environment
load_dotenv()
os.environ['LANGCHAIN_API_KEY'] = os.getenv('LANGCHAIN_API_KEY')
os.environ['LANGCHAIN_TRACING_V2'] = "true"
os.environ['LANGCHAIN_PROJECT'] = os.getenv('LANGCHAIN_PROJECT')
```

```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://www.xevensolutions.com/")
loader
```

```python
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000 , chunk_overlap = 200)
document = text_splitter.split_documents(docs)
document
```

```python
from langchain_community.llms import Ollama
llm = Ollama(model = 'llama3.2:1b')
```

```python
os.environ['HUGGINGFACE_API_KEY'] = os.getenv("HUGGINGFACE_API_KEY")
from langchain_huggingface import HuggingFaceEmbeddings
embeddings2 = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
from langchain_community.vectorstores import FAISS
vectorstore = FAISS.from_documents(document , embeddings2)
```

```python
query = "What are the Services of Xeven Solutions"
response = vectorstore.similarity_search(query)
response[0].page_content
```

```python
## Retrieval Chain , Document Chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template(
    """
Anwer the following question based only on the provided context:
<context>
{context}
</context>
"""
)
document_chain = create_stuff_documents_chain(llm , prompt)
document_chain
```

## **`Making VectorDataBase as Retriever`**

```python
retriever = vectorstore.as_retriever()
retriever
```

```python
from langchain.chains import create_retrieval_chain
retreival_chain = create_retrieval_chain(retriever , document_chain) # now it will get its context on its own we dont need to provide specific context to the model to give us correct information
```

```python
# get the response from the LLM 
response = retreival_chain.invoke({
    "input" : "What type of services do Xeven Solutions offer?"
})
```
